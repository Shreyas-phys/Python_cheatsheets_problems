```python
creators={"Python": "Guido van Rossum", "C": "Dennis Ritchie", "Java": "James Gosling"}

for element in creators:
    print(f"{element} was created by {creators[element]}")

# more pythonic
for lang, author in creators.items():
    print(f"{lang} was created by {author}")
```

```python
evens=set()

for i in range(1,11):
    evens.add(i)

print(evens)

primes={2, 3, 5, 7}

#intersection between the two sets
print(evens | primes)   # union        {1, 2, 3, 4, 5, 6}
print(evens & primes)   # intersection {3, 4}
print(evens - primes)   # difference   {1, 2}
print(evens ^ primes)   # symmetric difference {1, 2, 5, 6}
```
