## 2. Lists & Indexing <a id="slicing"></a>

Lists are ordered, mutable collections.


We can put other types of things in lists:


A list can contain a mix of different types of variables:

```python

```


| Operation | Meaning         | Output |
| --------- | --------------- | ------ |
| `arr[0]`  | first element   | `10`   |
| `arr[1]`  | second element  | `20`   |
| `arr[-1]` | last element    | `50`   |
| `arr[-2]` | second from end | `40`   |




Negative indexing is simply counting backward from the end.

## 2.2 Basic Slicing (start : stop : step)

### Core Syntax
arr[start : stop]  
arr[start : stop : step]

### Basic Slices
```python
arr[0:3]      # [10, 20, 30]
arr[2:5]      # [30, 40, 50]

arr[:3]       # [10, 20, 30]   (start = 0)
arr[2:]       # [30, 40, 50]   (stop = end)
```

### Step Slices
```python
arr[::2]      # [10, 30, 50]
arr[1::2]     # [20, 40]
arr[::-1]     # [50, 40, 30, 20, 10]
arr[::-2]     # [50, 30, 10]
```
### Slicing with Negative Indices
```python
arr[-4:-1]    # [20, 30, 40]
arr[-4:]      # [20, 30, 40, 50]
arr[:-2]      # [10, 20, 30]

arr[-1:0:-1]  # [50, 40, 30, 20]
```
### Bounds Never Error (Important in ML loops)

```python
arr[0:999]    # OK
arr[-999:3]   # OK
```
Python returns what exists.

But regular indexing DOES error


```python
arr[999]      # âŒ IndexError
```

### List Methods <a id="list_methods"></a>



````python
nums=[1,23,45,67]
# append() â€” add an item at the end
nums.append(10)
print(nums)

# list.pop removes and returns the last element of a list:
nums.pop()
print(nums)

# extend() â€” append elements from another list
# extend([20, 30]) adds two elements
nums.extend([20, 30])
print(nums)

# append([20, 30]) adds one element (a list)
nums.append([40, 50])
print(nums)

# insert(index, value) â€” insert at a position
nums.insert(2, 99)
print(nums)

nums.pop()
print(nums)

# pop(index) â€” remove and return element
nums.pop(2)
print(nums)

nums.extend([1,1,1,1,1])
print(nums)
# count() â€” count occurrences of a value
print(nums.count(1))  # returns 5


# remove(value) â€” remove first matching value
nums.remove(1)

## Searching lists:
nums.index(23)  # returns the index of the first occurrence of 23
nums.index(100)  # throws an error because 100 is not in the list
````

To avoid unpleasant surprises like this, we can use the in operator to determine whether a list contains a particular value:

````python
# Is Earth a planet?
print("Earth" in planets)
# Is 22 in nums?
print(22 in nums)
````

There are a few more interesting list methods we haven't covered. If you want to learn about all the methods and attributes attached to a particular object, we can call help() on the object itself. For example, help(planets) will tell us about all the list methods

# 3. Tuples <a id="tuples"></a>

Tuples are almost exactly the same as lists. They differ in just two ways.

1: The syntax for creating them uses parentheses instead of square brackets.

2: They cannot be modified (they are immutable).

Tuples are often used for functions that have multiple return values.

For example, the as_integer_ratio() method of float objects returns a numerator and a denominator in the form of a tuple:

````python
t = (1, 2, 3)
print(t[0])  # prints 1

# t[0] = 10  # throws an error because tuples are immutable
t[0]=10
````

# 4. Loops and List Comprehensions <a id="lists"></a>
This section focuses on the **two most important loop tools** in Python:
- `range()`  
- `enumerate()`  

`range()` produces a sequence of integers, typically used in `for` loops.

### Syntax:
range(start, stop, step)
### Notes:
- `start` is inclusive  
- `stop` is **exclusive** (just like slicing)  
- `step` is optional (defaults to 1)

````python
# Basic ranges
list(range(5))           # [0,1,2,3,4]
list(range(1, 6))        # [1,2,3,4,5]
list(range(2, 10, 2))    # [2,4,6,8]

# Negative step (reverse range)
list(range(10, 0, -1))   # [10,9,8,...,1]
````

```python
for index, value in enumerate(iterable):
```
### Why it's important:
- Cleaner than using `range(len(arr))`
- Very common in ML code when looping through batches

## When to use enumerate() vs range()

### âœ” Use `enumerate()` when:
- You want **index + value**
- You care about **readability**

### âœ” Use `range()` when:
- You ONLY need indices
- Youâ€™re looping a fixed number of times (not tied to a list)


````python
names = ["Amy", "Bob", "Charlie"]

for idx, name in enumerate(names):
    print(idx, name)

# With a custom starting index:
for idx, name in enumerate(names, start=1):
    print(idx, name)
````

# 5. Loops and List Comprehensions <a id="list_comprehension"></a>
 
List comprehensions are one of the most useful Python features.
They allow you to build lists **in a single, clear, fast expression**.

General form:
```python
[new_element  for  variable in iterable   if condition]
```

## ðŸ”¥ List Comprehension Quick Syntax Reference

### 1) Transform
```python
[new for x in arr]
```
### 2) Filtering ( if statement goes at the END. ALWAYS goes after the for. It is filtering)
```python
[x for x in arr if condition]
```
### 3) Transform + Filter
```python
[new for x in arr if condition]
```
### 4) Ifâ€“Else inside (goes at the START)
```python
VALUE if CONDITION else VALUE2  for x in ...

```
### 5) Nested loops (rare)
```python
[x for row in matrix for x in row]
```

````python
# They replace loops like
squares = []
for x in nums:
    squares.append(x*x)
print(squares)

#with list comprehensions, we can write this as
squares = [x*x for x in nums]
print(squares)
#The variable after for is the same variable you must use inside the expression.


# with an optional condition:
even_squares = [x*x for x in nums if x % 2 == 0] # only even numbers
print(even_squares)

#Conditional Transformation
labeled_nums = ['even' if x % 2 == 0 else 'odd' for x in nums]
print(labeled_nums)


#Nested Loops (Flattening)
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flattened = [num for row in matrix for num in row]
print(flattened)
#Order: left-to-right = nested loops.


# with functions
def f(x):
    return x*x + 1

vals = [f(x) for x in nums]
print(vals)


#List Comprehensions + Strings
chars = [c for c in "hello"]          # ['h','e','l','l','o']
uppercased = [word.upper() for word in ["a","b","c"]]
print(chars)
print(uppercased)
````

# 6. Strings <a id="strings"></a>

This section will cover some of Python's built-in string methods and formatting operations.

Strings are sequences
Strings can be thought of as sequences of characters. Almost everything we've seen that we can do to a list, we can also do to a string.

Such string manipulation patterns come up often in the context of data science work.
strings in Python can be defined using either single or double quotations. They are functionally equivalent.

````python
x = 'Pluto is a planet'
y = "Pluto is a planet"
x == y
````

Double quotes are convenient if your string contains a single quote character (e.g. representing an apostrophe).

Similarly, it's easy to create a string that contains double-quotes if you wrap it in single quotes:

```python
print("Pluto's a planet!")
print('My dog is named "Pluto"')
```
If we try to put a single quote character inside a single-quoted string, Python gets confused:

```python
'Pluto's a planet!'
```
We can fix this by "escaping" the single quote with a backslash.

```python
'Pluto\'s a planet!\n'
```
The last sequence, \n, represents the newline character. It causes Python to start a new line.

In addition, Python's triple quote syntax for strings lets us include newlines literally (i.e. by just hitting 'Enter' on our keyboard, rather than using the special '\n' sequence). We've already seen this in the docstrings we use to document our functions, but we can use them anywhere we want to define a string.

````python
hello = "hello\nworld"
print(hello)

triplequoted_hello = """hello
world"""
print(triplequoted_hello)
triplequoted_hello == hello
````
